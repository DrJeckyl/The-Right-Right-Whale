```python
%pylab inline
import graphlab as gl # Graph lab suite
import os
import sys
import shutil # For managing files and storage
```

```python
# Read in the train.csv data as an SFrame (sort of like a pandas/R data frame)
trainingSet = gl.SFrame.read_csv('train.csv')
```

```python
# Data is formatted as:
trainingSet.head()
```

```python
# Make a list of all the unique whale ID's
uniqueWhales = trainingSet['whaleID'].unique()

# Dictionary to map img name to whale ID -- sort of reverse of train.csv
whaleIDMap = dict([(i, val) for i, val in zip(trainingSet['Image'], trainingSet['whaleID'])])
```

```python
# Directory where the cropped training images are
trainingImgDir = os.getcwd() + "\\RecognitionTrainingImages"
```

```python
try:
    # Try to load the sframe from disk
    trainSF = gl.SFrame('./sframe/trainShuffle')
except:
    # Create the sframe from scratch, then save it
    # Load images from trainingImgDir in a random order
    trainSF = gl.image_analysis.load_images(trainingImgDir, random_order=True)
    # Add a column analogous to whaleID in trainingSet
    # Uses the whaleIDMap and the path name to get the whaleID
    # x.split('/')[-1] should give the name of the image, e.x. 'w_7812.jpg'
    trainSF['whaleID'] = trainSF['path'].apply(lambda x: whaleIDMap[x.split('/')[-1]])
    trainSF.save('./sframe/trainShuffle')
```

```python
# trainSF now looks like:
trainSF.head()
```

```python
# Not sure exactly what this does, but it was in the example
gl.canvas.set_target('ipynb')
```

```python
# Load the pretrained imagenet model
pretrainedModel = gl.load_model('http://s3.amazonaws.com/GraphLab-Datasets/deeplearning/imagenet_model_iter45')
```

```python
# Add another column to train, which is the extracted features from the pretrained model
# This step can take a while
trainSF['features'] = pretrainedModel.extract_features(trainSF)
```

```python
# Use graphlabs classifer to find the best model
# Can also take a long time
simpleClassifier = graphlab.classifier.create(trainSF, features = ['features'], target = 'whaleID')
```

```python
simpleClassifier.predict_topk(trainSF, output_type='probability')
```

```python
RandomForest = gl.random_forest_classifier.create(trainSF, features=['features'], target = 'whaleID')
```

```python
BoostedTrees = gl.boosted_trees_classifier.create(trainSF, features=['features'], target = 'whaleID')
```

## Now for the submission output

```python
# Store the predictions in this SFrame
predictions = Logistic.predict_topk(trainSF, output_type='probability', k=10)
```

```python
# Quick dictionary to give the index position of each whaleID
labels = list(uniqueWhales)
vectorMap = dict([i[::-1] for i in enumerate(labels)])
```

```python
# For testing
imagesToPredict = trainSF

probabilityVector = []
for i in range(len(imagesToPredict['path'])):
    # Initialize the output vector to be the size of the number of unique whales
    out_vector = zeros(len(labels))
    # Set the value of the probability for each whaleID
    # Couldn't figure out a cleaner way to do this
    
    # A list of the indexes where there are probabilities defined
    # vectorMar gives the index location of the label found from the filter operation
    indices = [vectorMap[j['class']] for j in predictions.filter_by(i, 'id')]
    
    # The assignment is a same sized list of the probabilities given by the same filter operation
    out_vector[indices] = [j['probability'] for j in predictions.filter_by(i, 'id')]
    
    # Add this row to the output
    probabilityVector.append(out_vector)
```

```python
probabilityVector = array(probabilityVector)
# Make a new data frame which will be exported as csv
# The first colum is the image name followed by the probability vector
submission = gl.SFrame({'Image':[i.split('/')[-1] for i in imagesToPredict['path']]})
for i in range(len(labels)):
    submission.add_column(gl.SArray(probabilityVector[:,i]), name=labels[i])
    
```
